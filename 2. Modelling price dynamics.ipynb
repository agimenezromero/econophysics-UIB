# Generate random numbers distributed according to a particular distribution

## Uniform distribution

The probability density function of the **uniform** distribution is given by

\begin{equation}
f(x)= \begin{cases}\frac{1}{b-a} & \text { for } a \leq x \leq b \\ 0 & \text { for } x<a \text { or } x>b\end{cases}
\end{equation}

To generate uniformly distributed random numbers we can use the **numpy.random.uniform(a,b, N)** function


```python
a = 0
b = 1

N = 10**6

u = np.random.uniform(a, b, N)

plt.hist(u, bins=100, density=True);
```

## Gaussian distribution

The probability density function of the **Gaussian** distribution is given by

\begin{equation}
f(x)=\frac{1}{\sigma \sqrt{2 \pi}} \exp\left[-\frac{1}{2}\left(\frac{x-\mu}{\sigma}\right)^2\right]
\end{equation}

where $\mu$ is the mean and $\sigma$ is the standard deviation.

To generate uniformly distributed random numbers we can use the **numpy.random.normal(μ,σ, N)** function


```python
def gaussian(x, mu, sigma):

  return 1/(sigma*np.sqrt(2*np.pi)) * np.exp(-(x-mu)**2/(2*sigma**2))

mu = 0
sigma = 1

N = 5*10**6

n = np.random.normal(mu, sigma, N)

x = np.linspace(np.amin(n), np.amax(n), 1000)

plt.hist(n, bins=100, density=True)
plt.plot(x, gaussian(x, mu, sigma), lw=3, color="r")
```

## Exponential distribution

The probability density function of the **exponential** distribution is given by

\begin{equation}
f(x)=\exp (-x)
\end{equation}

but it is usually written as

\begin{equation}
f(x; \lambda)=\lambda \exp (-\lambda x)
\end{equation}

Both formulations are related by

\begin{equation}
f(x; \lambda)=f\lambda(\lambda x)=\lambda\exp (-\lambda x)
\end{equation}

To generate exponentially distributed random numbers we can use the **scipy.stats.expon(size=N)** function. Furthermore, the pdf can be easily plotted using **stats.expon.pdf(x)** function.


```python
N = 5*10**6

e = stats.expon.rvs(size=N)

x = np.linspace(np.amin(e), np.amax(e), 1000)

plt.hist(e, bins=100, density=True)
plt.plot(x, stats.expon.pdf(x), lw=3, color="r")
```

## Beta distribution

The probability density function of the **Beta** distribution is given by

\begin{equation}
f(x; a, b)=\frac{\Gamma(a+b) x^{a-1}(1-x)^{b-1}}{\Gamma(a) \Gamma(b)} \,
\end{equation}

where $a$ and $b$ are shape parameters and $\Gamma$ is the gamma function.

To generate exponentially distributed random numbers we can use the **scipy.stats.beta(a, b, size=N)** function. Furthermore, the pdf can be easily plotted using **stats.beta.pdf(x, a, b)** function.


```python
a = 2
b = 5

N = 5*10**6

beta = stats.beta.rvs(a, b, size=N)

x = np.linspace(np.amin(beta), np.amax(beta), 1000)

plt.hist(beta, bins=100, density=True)
plt.plot(x, stats.beta.pdf(x, a, b), lw=3, color="r")
```

## Gamma distribution

The probability density function of the **Gamma** distribution is given by

\begin{equation}
f(x; a)=\frac{x^{a-1} e^{-x}}{\Gamma(a)}
\end{equation},

where $a$ is the shape parameter. It can also be found as

\begin{equation}
f(x ; \alpha, \beta)=\frac{x^{\alpha-1} e^{-\beta x} \beta^\alpha}{\Gamma(\alpha)} \ ,
\end{equation}

where $\alpha$ and $\beta$ are denoted as the shape and rate parameters.

Both formulations are related by

\begin{equation}
f(x ; \alpha, \beta)=\beta \cdot f(\beta x; a)=\beta\frac{\beta^{\alpha-1}x^{\alpha-1}e^{-\beta x}}{\Gamma{a}}=\frac{x^{\alpha-1} e^{-\beta x} \beta^\alpha}{\Gamma(\alpha)}
\end{equation}

```python
a = 2

N = 5*10**6

gamma = stats.gamma.rvs(a, size=N)

x = np.linspace(np.amin(gamma), np.amax(gamma), 1000)

plt.hist(gamma, bins=100, density=True)
plt.plot(x, stats.gamma.pdf(x, a), lw=3, color="r")
```

## Student's t distribution

The probability density function of the Student's t distribution is given by

\begin{equation}
f(x; \nu)=\frac{\Gamma((\nu+1) / 2)}{\sqrt{\pi \nu} \Gamma(\nu / 2)}\left(1+x^2 / \nu\right)^{-(\nu+1) / 2} \ ,
\end{equation}

where $df$ is the number of degrees of freedom.

```python
df = 10

N = 5*10**6

t = stats.t.rvs(df, size=N)

x = np.linspace(np.amin(t), np.amax(t), 1000)

plt.hist(t, bins=100, density=True)
plt.plot(x, stats.t.pdf(x, df), lw=3, color="r")
```

Note that the limit of the Student's t distribution when the degrees f freedom go to infinite is the normal distribution



```python
df_1 = 5
df_2 = 10
df_3 = 50
df_4 = 5000


x = np.linspace(-5, 5, 1000)

plt.figure(figsize=(8, 6))

plt.plot(x, stats.t.pdf(x, df_1), lw=3, color="C0", label="df=%i" % df_1)
plt.plot(x, stats.t.pdf(x, df_2), lw=3, color="C1", label="df=%i" % df_2)
plt.plot(x, stats.t.pdf(x, df_3), lw=3, color="C2", label="df=%i" % df_3)
plt.plot(x, stats.t.pdf(x, df_4), lw=3, color="C3", label="df=%i" % df_4)

plt.plot(x, gaussian(x, 0, 1), color="k", ls="--", lw=3, label="N(0,1)")

plt.yscale("log")

plt.legend();
```

#Fitting a data sample to a distribution

To fit a particular distribution to a given data sample we can use the **scipy.stats.fit** function, which used the Maximum Likelihood Estimation (MLE) method to estimate the parameters of the distribution.

It works both with continuous and discrete distributions, estimating their probability density function (pdf) and probability mass function (pmf), respectively.

```python
#Generate some random numbers according to a Binomial distribution
n = 10
p = 0.3

N = 10**5

r = stats.binom.rvs(n, p, size=N)

#Generate some random numbers according to a beta distribution
a = 1
b = 5

a2 = 5
b2 = 3

N = 10**5

r2 = stats.beta.rvs(a,b, size=N)
r3 = stats.beta.rvs(a2,b2, size=N)

#Fit the samples to the binomial and beta distributions
shape_bounds = {'n': (0, 20), 'p': (0, 1)}
shape_bounds_2 = {'a': (0, 10), 'b':(0, 10)}

fit = stats.fit(stats.binom, r, shape_bounds)
fit2 = stats.fit(stats.beta, r2, shape_bounds_2)
fit3 = stats.fit(stats.beta, r3, shape_bounds_2)

print("---- Fit of Binomial distribution ----\n")
print(fit)

print("\n ---- Fit of beta distribution ----\n")
print(fit2)

print("\n ---- Fit of beta distribution (2) ----\n")
print(fit3)
print()

fig, ax = plt.subplot_mosaic("""AB""", figsize=(8*2, 6))

n_fit, p_fit = fit.params[0], fit.params[1]

x = np.arange(0, 11, 1)

ax["A"].hist(r, bins=np.arange(0, n+2, 1), density=True);
ax["A"].plot(x, stats.binom.pmf(x, n_fit, p_fit), ls="-", marker="o", color="r", ms=10)

ax["A"].set_title("Fit to a Binomial distribution")
ax["A"].set_xlabel("x")
ax["A"].set_ylabel("PDF")

a_fit, b_fit = fit2.params[0], fit2.params[1]
a2_fit, b2_fit = fit3.params[0], fit3.params[1]

x2 = np.linspace(0, 1, 100)

ax["B"].hist(r2, bins=100, density=True);
ax["B"].plot(x2, stats.beta.pdf(x2, a_fit, b_fit), color="r", lw=3)

ax["B"].hist(r3, bins=100, density=True);
ax["B"].plot(x2, stats.beta.pdf(x2, a2_fit, b2_fit), color="r", lw=3)

ax["B"].set_title("Fit to a Beta distribution")

ax["B"].set_xlabel("x")
ax["B"].set_ylabel("PDF")
```
